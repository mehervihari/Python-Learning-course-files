```
# create a gameboard class
class GameBoard:
# create instance of this class
    def __init__(self):
        self.game_board = {1:' ', 2:' ', 3:' ', 4:' ', 5:' ', 6:' ', 7:' ', 8:' ', 9:' '}
# create item as item setter function
    def set_items(self, user, position, game_board):
        game_board[position] = user
        return game_board    
# decorator for the gameboard function to add separate self instance for game_board            
    @property
    def gameBoard(self):
        return self.game_board
# add another function called clearboard
    def clearBoard(self):
        self.game_board = {1:' ', 2:' ', 3:' ', 4:' ', 5:' ', 6:' ', 7:' ', 8:' ', 9:' '}
# add another function called is_Take_Place
    def is_Place_Taken(self, game_board, index):
        if game_board[index] != ' ':
            return True
        return False
# add another function called is_Board_Full            
    def is_Board_Full(self, game_board):
        for index in range(1, 10):
            if game_board[index] == ' ':
                return False
        return True        
# add another function called is_Game_Won
    def is_Game_Won(self, game_board):
        win_conds = ((1,2,3),(4,5,6),(7,8,9),(1,4,7),(2,5,8),(3,6,9),(1,5,9),(3,5,7))
        for win_cond in win_conds:
            if game_board[win_cond[0]] == game_board[win_cond[1]] and game_board[win_cond[1]] == game_board[win_cond[2]] and game_board[win_cond[0]] != ' ':
                return True
        return False
# add another function to print gameboard: contains if statement to control game flow in the board:
    def printGameBoard(self, game_board):
        index = 0
        for row in range(1, 4):
            for column in range(1, 4):
                index += 1
                if column != 3:
                    print(game_board[index], end = '')
                    print('|', end='')
                else: print(game_board[index])


# create a 2nd class called Game: to control the game start and game end, ask player names

class Game:

#1 - game start
    def game_start(self):

        self.controlBoard = GameBoard()
        self.game_board = self.controlBoard.gameBoard
        self.playerOne = '0'
        self.playerTwo = 'X'
        print("welcome to X-0 game")
        print("please enter player1's name")
        self.player_one = input(' : ')
        print("please enter player2's name")
        self.player_two = input(' : ')
        print("welcome to X-0 game, each place is represented by 1-9, starting from left column each time and moving along the row")
        self.controlBoard.printGameBoard(self.game_board)
        self.turn = 1
#2 - end the game and play again
    def game_end(self):
        # checking if player wants to end the game   
        if self.game_running == False:
            replay = input("press 0 to quit or 1 for replay")
            try:
                if int(replay):
                    self.game_running = True
                    self.game_start()
            except:
                print("a number must be entered")
                self.game_end()
#3 - game turn
    def takeTurn(self, user, item):
        print(user + " choose a place between 1-9")
        try:
            position = int(input(': '))
            if position > 9 or position < 1:
                raise Exception
        except:
            print("pick a number between 1-9")
            return self.takeTurn(user, item)

        if self.controlBoard.is_Place_Taken(self.game_board, position):
            print("that place is taken")
            self.takeTurn(user, item)
        else:
            self.controlBoard.set_items(item, position, self.game_board)
            self.controlBoard.printGameBoard(self.game_board)
            if (self.controlBoard.is_Game_Won(self.game_board)):
                print(user + " wins")
                self.game_running = False
#4 - game manager
    def main(self):
        self.game_running = True
        self.game_start()
        while self.game_running:
            if self.turn%2 != 0:
                self.takeTurn(self.player_one, 'X')
            else: self.takeTurn(self.player_two, '0')

            if self.controlBoard.is_Board_Full(self.game_board):
                print("its a draw, no one wins")
                self.game_running = False
            self.turn += 1

            if not self.game_running:
                self.game_end()
                
# create the launcher for the game
if __name__ == '__main__':
    Game().main()
            
```
