```
# import the required python module
import random

# pick and save the words to be guessed in hangman_guessing.py file, as a list
from hangman_guessing import guess_list

# use the imported python module to choose a random word from the imported list of words you saved
guess_word = random.choice(guess_list)
word_length = len(guess_word)

# keep the game running until the game is over
game_over = False

# apply tries to the players to use try to win
tries = 6

# save game shapes in hangman_life.py file and import those files from this file
from hangman_life import game_name
print(game_name)
#print(f'the word you guessed is {guess_word}')
# create a list containing 6 '_' underscores to store and save the right letters guessed by the user
result = []

# iterate over the player's input
for _ in range(word_length):
    result += '_'

while not game_over:
    user_guess = input('Guess a letter: ')

# inform the player about the letter they have guessed after they did
    if user_guess in result:
        print(f'the letter you have guessed is: {user_guess}')

# check if the letter guessed by the user is right or wrong
    for position in range(word_length):
        letter = guess_word[position]   
        if letter == user_guess:
            result[position] = letter

# player loses a try if the guessed letter is worng
    if user_guess not in guess_word:
        print(f'you guessed {user_guess}, this letter is not present in the word, you lose a try')

# if the players lose all their tries, end the game and print the game over message
        tries -= 1
        if tries == 0:
            game_over = true
            print('Game over, You lose the game, try again')          

# join all the letters that are present in the list of guessed letters and turn it into string
    print(f"{' '.join(result)}")

# check if the player guessed all the letters right to print the win message
    if '_' not in result:
        game_over = True
        print('you are a winner, congrats')

# print the correct shape after every player's try
    from hangman_life import lives
    print(lives[tries])    
```
